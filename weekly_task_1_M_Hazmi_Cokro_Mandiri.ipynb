####18. Perbaiki program dibawah sehingga menghasilkan output yang sesuai 

```
# Fungsi Move
def move(n , x, y, z):
    if n == 1:
        print("Pindah Kelas {} ke {}".format(x,y))
        return
    move(n-1, x, z, y)
    print("Pindah Kelas {} ke {}".format(x,y))
    move(n-1, z, y, x)

move(3, 'A', 'C', 'B') 

# Output :
# Pindah Kelas A ke C
# Pindah Kelas A ke B
# Pindah Kelas C ke B
# Pindah Kelas A ke C
# Pindah Kelas B ke A
# Pindah Kelas B ke C
# Pindah Kelas A ke C
```

####19. Tulis program Python untuk mencetak angka genap dari daftar tertentu. 

```
# input [1, 2, 3, 4, 5, 6, 7, 8, 9]
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Fungsi Lambda 
genap = list(filter(lambda x: (x % 2 == 0), a))

# Output [2, 4, 6, 8]
print(f'List Angka Genap = {genap}')
```
